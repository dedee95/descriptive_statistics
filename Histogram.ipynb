# Histogram

One of the best ways to present the distribution of data is with a histogram. A histogram is a good form of bar chart used to view the frequency distribution of a continuous data point. It plots two numerical axes, one axis is used for the number of occurrences of a particular value (frequency) and the other axis is used to group the values into bins (interval classes).

Each bar in the histogram represents the number of frequencies in each interval/class/bin. Histograms have several uses, including helping us give an estimate of where the values are concentrated, helping us see if there are any outlier values, and helping us give an institution about the type of distribution of our data.

Create histogram

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# create dummy dataset
age = 10 * np.random.randn(1000) + 30
age
```

### 1. Using Matplotlib

```python
# create simple histogram
plt.hist(age)
```

```python
# create histogram with parameter adjustment
plt.figure(figsize=(9,7))
plt.hist(age, bins=20, color='yellowgreen', edgecolor='black', alpha=0.7)
plt.title('Histogram', {'fontsize':22})
plt.ylabel('Frequency', {'fontsize':14})
plt.xlabel('Age', {'fontsize':14})
plt.grid(axis="y", which='minor')
plt.show()
```

### 2. Using Seaborn

```python
sns.histplot(x=age)
```

```python
sns.set_style("ticks")
plt.figure(figsize=(9,7))
sns.histplot(x=age, bins=11, kde=True, element="step")
plt.title('Histogram', {'fontsize':22})
plt.ylabel('Frequency', {'fontsize':14})
plt.xlabel('Age', {'fontsize':14})
plt.show()
```

### 3. Using PLotly

```python
fig = px.histogram(x=age)
fig.show()
```

```python
fig = px.histogram(x=age, nbins=15)
fig.update_layout(title = 'Histogram',
                  title_x = 0.48,
                  title_font= dict(size = 24, color = '#7C7C7C' ))
fig.update_xaxes(title= 'Age')
fig.update_yaxes(title= 'Frequency')
fig.show()
```
